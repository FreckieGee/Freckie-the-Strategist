## Facial Detection using Opencv

```python
#import Library
import cv2
```

```python
#Step 1: Identify on Webcam
capture=cv2.VideoCapture(0) # 0 - for local Webcam and 1 for External Webcam
```

```python
#Step 3 Leverage Haarcascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_eye.xml")
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_smile.xml")
```

```python
#Step 4: Create a function for facial detection
def detect(gray,frame):
    faces=face_cascade.detectMultiScale(gray,1.3, 5)
    for(x,y,w,h) in faces:
        cv2.rectangle(frame, (x,y),(x+w, y+h),(255,0,0),5)
        roi_gray=gray[y:y+h, x:x+w]
        roi_color=frame[y:y+h, x:x+w]
        
        #Eyes detection programme
        eyes=eye_cascade.detectMultiScale(roi_gray, 1.1, 3)
        for(ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color, (ex,ey),(ex+ew, ey+eh),(0,0,255),3)
            
         #Smile detection programme
        smile = smile_cascade.detectMultiScale(roi_gray, 1.8, 25)
        for(sx,sy,sw,sh) in smile:
            cv2.rectangle(roi_color, (sx,sy),(sx+sw, sy+sh),(7,139,255),3)
        
        
    return frame
```

```python
# Step 2: Switch on Webcam

while True:
    _ , frame = capture.read()
    
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    display=detect(gray,frame)
    cv2.imshow("Video",frame)
    if cv2.waitKey(1) & 0xff==ord("q"):
        break
capture.release()
cv2.destroyAllWindows()
```
